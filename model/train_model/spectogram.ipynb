```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# โหลดไฟล์เสียง
audio_path = r'C:\Users\iluvpham\Documents\GitHub\DSP_project\ignoredir\data_test\guitar-solo-5999.wav'  # ใส่ path ของไฟล์เสียงที่คุณต้องการใช้
y, sr = librosa.load(audio_path)

# สร้าง Spectrogram และ Mel-Spectrogram
D = np.abs(librosa.stft(y))**2  # คำนวณ Spectrogram (พลังงาน)
S = librosa.feature.melspectrogram(S=D, sr=sr)  # แปลงเป็น Mel-Spectrogram
S_dB = librosa.power_to_db(S, ref=np.max)  # แปลงเป็น Decibel Scale
```

```python
# สร้างกราฟ
plt.figure(figsize=(12, 8))

# Plot Waveform
plt.subplot(3, 1, 1)
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform')
plt.xlabel('Time')
plt.ylabel('Amplitude')

# Plot Spectrogram
plt.subplot(3, 1, 2)
librosa.display.specshow(librosa.amplitude_to_db(D, ref=np.max), sr=sr, x_axis='time', y_axis='linear', cmap='magma')
plt.title('Spectrogram')
plt.xlabel('Time')
plt.ylabel('Frequency')

# Plot Mel-Spectrogram
plt.subplot(3, 1, 3)
librosa.display.specshow(S_dB, sr=sr, x_axis='time', y_axis='mel', cmap='viridis')
plt.title('Mel-Spectrogram')
plt.xlabel('Time')
plt.ylabel('Mel Filters')

# ปรับระยะห่างระหว่างกราฟและแสดงผล
plt.tight_layout()
plt.show()
```
